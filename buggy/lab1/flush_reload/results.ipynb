```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

x = np.genfromtxt("indices.txt", dtype=int)
y = np.genfromtxt("access_times.txt", dtype=int)

# Print the most frequent int in Output.txt
most_frequent = np.bincount(x).argmax()
```

```python
# xpoints = np.array([1, 8])
# ypoints = np.array([3, 10])

plt.plot(x, y, "o")
plt.xlabel("Index that was Flushed+Reloaded")
plt.ylabel("Access Time (Cycles)")
plt.title("Flush+Reload Results (300 Samples)")
plt.xticks(np.array([most_frequent]))
plt.show()
```
